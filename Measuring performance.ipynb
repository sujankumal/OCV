```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import time
```

```python
img1 = cv2.imread('nature.jpg')
e1 = time.time()
for i in range(5,49,2):
    img1 = cv2.medianBlur(img1,i)
plt.imshow(img1)
plt.show()
e2 = time.time()
tf = cv2.getTickFrequency()
t = (e2 - e1)/tf
print(e1,e2,tf, t)
```

```python
img1 = cv2.imread('image1.jpg')
e1 = cv2.getTickCount()
for i in range(5,49,2):
    img1 = cv2.medianBlur(img1,i)
plt.imshow(img1)
plt.show()
e2 = cv2.getTickCount()
tf = cv2.getTickFrequency()
t = (e2 - e1)/tf
print(e1,e2,tf, t)
```
